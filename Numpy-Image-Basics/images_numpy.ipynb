```python
plt.imshow(pic_arr) # RGB
```

```python
pic_red = pic_arr.copy()
pic_green = pic_arr.copy()
pic_blue = pic_arr.copy()
```

```python
# R, G, B
# Closer to white when more red exists
pic_red = pic_red[:,:,0]
plt.imshow(pic_red, cmap='gray')
```

```python
# R, G, B
pic_green = pic_green[:,:,1]
plt.imshow(pic_green, cmap='gray')
```

```python
# R, G, B
pic_blue = pic_blue[:,:,2]
plt.imshow(pic_blue, cmap='gray')
```

```python
remove_green = pic_arr.copy()
remove_green[:,:,1] = 0
```

```python
# Looks purple because only red and blue exist
plt.imshow(remove_green)
```

```python
remove_blue = remove_green.copy()
remove_blue[:,:,2] = 0
```

```python
# Only red remains
# Actually red because its showing all three channels
plt.imshow(remove_blue)
print(remove_blue.shape)
```
