```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
orig = cv2.imread('../DATA/00-puppy.jpg')
plt.imshow(orig)
```

```python
rgb_img = cv2.cvtColor(orig, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
hsv_img = cv2.cvtColor(orig, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
```

```python
hls_img = cv2.cvtColor(orig, cv2.COLOR_RGB2HLS)
plt.imshow(hls_img)
```
