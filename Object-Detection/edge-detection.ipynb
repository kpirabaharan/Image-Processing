# Edge Detection

## Using Canny Method

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/sammy_face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
med_val = np.median(img)
med_val
```

```python
# Lower threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0, 0.7 * med_val))
# Upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255, 1.3 * med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
blur = cv2.blur(img, ksize=(7, 7))
edges = cv2.Canny(image=blur, threshold1=lower, threshold2=upper+50)
plt.figure(figsize=(16, 12))
plt.subplot(121)
plt.imshow(blur)
plt.subplot(122)
plt.imshow(edges)
```
