# Contour Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/internal_external.png', 0)
plt.imshow(img, cmap='gray')
```

```python
contours, hierarchy  = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
len(image)
```

## External Contours

```python
external_contours = np.zeros(img.shape)
external_contours.shape
plt.imshow(img)
```

```python
for i in range(len(contours)):
    # EXTERNAL CONTOURS
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours, contours, i, 255, -1)
```

```python
plt.imshow(external_contours, cmap='gray')
```

## Internal Contours

```python
internal_contours = np.zeros(img.shape)
internal_contours_1 = np.zeros(img.shape)
internal_contours_2 = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    # INTERNAL CONTOURS
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contours, contours, i, 255, -1)
plt.imshow(internal_contours, cmap='gray')
```

### Can separate internal contours by which external contour its in

```python
for i in range(len(contours)):
    # INTERNAL CONTOURS
    if hierarchy[0][i][3] == 0:
        cv2.drawContours(internal_contours_1, contours, i, 255, -1)
plt.imshow(internal_contours_1, cmap='gray')
```

```python
for i in range(len(contours)):
    # INTERNAL CONTOURS
    if hierarchy[0][i][3] == 4:
        cv2.drawContours(internal_contours_2, contours, i, 255, -1)
plt.imshow(internal_contours_2, cmap='gray')
```
